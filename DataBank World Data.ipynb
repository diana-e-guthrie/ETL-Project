```python
import pandas as pd
```

```python
data = pd.read_csv('World Bank GDP by Country data.csv')
data = data.dropna()
data
```

```python
data_df = data[["Series Code","Country Name","Country Code","2015 [YR2015]"]]
data_df.dropna
data_df
```

```python
gdpcode = "NY.GDP.MKTP.CD"
gdp = data_df[data_df["Series Code"]==gdpcode]
gdp
```

```python
totalpopcode = "SP.POP.TOTL"
totalpop = data_df[data_df["Series Code"]==totalpopcode]
totalpop.head(100)
```

```python
merge = gdp.merge(totalpop,left_on="Country Code",right_on="Country Code")
merge = merge.rename(columns={"2015 [YR2015]_x":"2015 GDP","2015 [YR2015]_y":"2015 Total Population","Country Name_x":"Country Name"})
merge = merge[["Country Name","Country Code","2015 GDP","2015 Total Population"]]
merge
```
